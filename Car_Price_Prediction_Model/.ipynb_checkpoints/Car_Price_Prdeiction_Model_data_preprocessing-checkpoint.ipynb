```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df_carproperties.columns = headers
```

```python
df_carproperties1 = df_carproperties.replace(["?",""," "], np.nan)
```

```python
df_carproperties1.isnull().sum()
```

#### Drop the missing values on the "price" column, since we are trying to predict the price.

```python
df_carproperties = df_carproperties1.dropna(subset=["price"], axis = 0)
df_carproperties
```

```python
df_carproperties.isnull().sum()
```

### The following columns have missing values
<ul>
    <li>"normalized-losses" has 37 missing values. fix: Replace the missing value with the mean</li>
    <li>"bore has" 4 missing values. fix: Replace the missing value with the mean</li>
    <li>"stroke" has 4 missing values. Replace the missing value with the mean</li>
    <li>"horsepower" has 2 missing values. Replace the missing value with the mean</li>
    <li>"peak-rpm" has 2 missing values. Replace the missing value with the mean</li>
</ul>

```python
df_carproperties.dtypes
```

### Calculating the mean of the columns where NaN is to be replaced

```python
avg_norm_loss = df_carproperties["normalized-losses"].astype("float").mean(axis=0)
avg_bore = df_carproperties["bore"].astype("float").mean(axis=0)
avg_stroke = df_carproperties["stroke"].astype("float").mean(axis=0)
avg_horsepower = df_carproperties["horsepower"].astype("float").mean(axis=0)
avg_peak_rpm = df_carproperties["peak-rpm"].astype("float").mean(axis=0)
```

#### Replacing the NaN for the respective columns with the column mean

```python
df_carproperties["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
df_carproperties["bore"].replace(np.nan, avg_bore, inplace = True)
df_carproperties["stroke"].replace(np.nan, avg_stroke, inplace = True)
df_carproperties["horsepower"].replace(np.nan, avg_horsepower, inplace = True)
df_carproperties["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace = True)
```

```python
df_carproperties
```

```python
df_carproperties["num-of-doors"].value_counts()
```

```python
df_carproperties["num-of-doors"].value_counts().idxmax()
```

```python
df_carproperties["num-of-doors"].replace(np.nan,'four', inplace = True)
```

```python
df_carproperties.isnull().sum()
```

```python
df_carproperties[["bore", "stroke", "price", "peak-rpm"]] = df_carproperties[["bore", "stroke", "price", "peak-rpm"]].astype("float")
df_carproperties[["normalized-losses", "horsepower"]] = df_carproperties[["normalized-losses", "horsepower"]].astype("int64")
```
