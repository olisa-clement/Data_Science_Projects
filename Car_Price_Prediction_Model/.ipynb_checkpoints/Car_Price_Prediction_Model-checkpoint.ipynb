```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
filename = "clean_df_carproperties.csv"
```

```python
df_carproperties = pd.read_csv(filename)
df_carproperties
```

```python
df_carproperties.corr()
```

```python
df_carproperties.corr()["price"].sort_values(ascending=False)
```

```python
predictors = ["engine-size", "curb-weight", "horsepower", "width", "highway-mpg", "length", "city-mpg"]
car_price = ["price"]
```

```python
df_carproperties[predictors].corr()
```

#### Creating a Linear regression object

```python
lm = LinearRegression()
```

#### fit the Linear Regresion model using the predictors

```python
lm.fit(df_carproperties[predictors], df_carproperties[car_price])
```

#### Obtain the intercept

```python
lm.intercept_
```

#### Obtain an array of correlation coeficients

```python
lm.fit(df_carproperties[predictors], df_carproperties[car_price]).coef_
```

```python
lm.get_params(deep=True)
```

### Calculating the R-square Value of the model

```python
print("The R-square value for the Regression model is: ", 
      round(lm.score(df_carproperties[predictors], df_carproperties[car_price]),4))
```
